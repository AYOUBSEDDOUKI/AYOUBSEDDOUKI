```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("/content/Marketing_Data.csv",sep=",")

x1 =  data.iloc[:,:1].values
x2 =  data.iloc[:,1:2].values
x3 =  data.iloc[:,2:3].values

x1 = (x1-np.mean(x1))/np.std(x1)
#print(x1)
#print("*******************************************************************")
x2 = (x2-np.mean(x2))/np.std(x2)
#print(x2)
#print("*******************************************************************")
x3 = (x3-np.mean(x3))/np.std(x3)
#print(x3)
#print("*******************************************************************")

x=x1
x = np.hstack((x,x2))
x = np.hstack((x,x3))
x = np.c_[np.ones(x.shape[0]),x]

'''
d=x[:,1:2]
d=np.hstack((d,x[:,2:3]))
d=np.hstack((d,x[:,3:4]))
d=np.hstack((d,x[:,0:1]))
'''

print(x)
print("*******************************************************************")
y =  data.iloc[:,3:4].values
y = (y-np.mean(y))/np.std(y)
```

```python
np.random.seed(0) # pour produire toujours le meme vecteur theta aléatoire
theta = np.random.randn(4, 1)
theta
```

```python
# defintion de model
def model(X,w):
    return X.dot(w)
```

```python
def cost_function(X, y, theta):
    return -2*np.sum((model(X, theta) - y)**2)
```

```python
cost_function(x, y, theta)
```

```python
def grad(X, y, theta):
    return X.T.dot(model(X, theta) - y)
```

```python
def gradient_descent(X, y, theta, learning_rate, n_iterations):
    
    cost_history = np.zeros(n_iterations) # création d'un tableau de stockage pour enregistrer l'évolution du Cout du modele
    
    for i in range(0, n_iterations):
        theta = theta - learning_rate * grad(X, y, theta) # mise a jour du parametre theta (formule du gradient descent)
        cost_history[i] = cost_function(X, y, theta) # on enregistre la valeur du Cout au tour i dans cost_history[i]
        
    return theta, cost_history
```

```python
n_iterations = 1000
learning_rate = 0.01

theta_final, cost_history = gradient_descent(x, y, theta, learning_rate, n_iterations)
```

```python
theta_final
```

```python
predictions = model(x, theta_final)
#print(predictions)
```

```python
from sklearn.metrics import r2_score as R2
r2 = R2(y,predictions)
print("le R2 :",r2)
n=len(x[:,0])
p=len(x[0,:])
R2_ajusté = 1 - ((1-r2)*(n-1))/(n-1-p)
print("le R2 ajusté:",R2_ajusté)
```
